```python
def estimate_jacobian(f : Callable[NDArray[np.float64], NDArray[np.float64]],
                     state : NDArray[np.float64], #1 dimensional vector
                      epsilon = 1000 #limit definition of the derivative applied to vectorvalued partials
                     ):
    y = f(state)
    n = state.shape[0]#function input vector shape
    m = y.shape[0]#function output vector shape
    J = np.zeros((m, n))
    for i in range(n):
        state_1 = np.copy(state)
        state_1[i] = state[i] + epsilon
        f_1 = f(state_1)
        for j in range(m):
            J[j][i] = (f_1[j] - y[j])/epsilon

    return J
```

```python

def raw_igdf_enu(height_from_sea_level : np.float64, lat : np.float64, lon : np.float64, time):
    #kilometers, degrees, degrees
    #returns B_east, b_north, b_up in nT: https://github.com/IAGA-VMOD/ppigrf/tree/main
    return ppigrf.igrf(lon, lat, height_from_sea_level, time)

def igdf_eci_vector(x : np.float64, y : np.float64, z : np.float64, time):
    #eci in METERS, returns unit eci igdf vector
    lat, lon, alt = pymap3d.eci2geodetic(x, y, z, time)
    b_east, b_north, b_up = raw_igdf_enu(alt/1000, lat, lon, time)
    ecefb_x, ecefb_y, ecefb_z = pymap3d.enu2ecef(b_east, b_north, b_up, lat, lon, alt)
    b_x, b_y, b_z = pymap3d.ecef2eci(ecefb_x, ecefb_y, ecefb_z, time)
    b_vec = np.array([b_x - x, b_y - y, b_z - z])
    return b_vec
    
    
```

```python
from astropy.time import Time
def right_ascension_to_longitude(right_ascension : np.float64, time = datetime.now()):
    #takes in degrees, returns values in degrees
    current_time = Time(time)
    current_sidereal = current_time.sidereal_time('apparent', 'greenwich')
    return (math.degrees(right_ascension) - current_sidereal.degree) % 360

def raw_sun_vector(time):
    #http://www.stargazing.net/kepler/sun.html
    #returns values in DEGREES
    date_relative_to_J2000 = (time - datetime(2000, 1, 1, 12, 0, 0)).days
    mean_longitude = (280.461 + 0.9856474 * date_relative_to_J2000) % 360
    mean_anomaly = (357.528 + 0.9856003 * date_relative_to_J2000) % 360

    ecliptic_longitude = mean_longitude + 1.915 * math.sin(math.radians(mean_anomaly)) + 0.020 * math.sin(2* math.radians(mean_anomaly))

    obliquity = 23.439 - 0.0000004 * date_relative_to_J2000
    Y = math.cos(math.radians(obliquity)) * math.sin(math.radians(ecliptic_longitude))
    X = math.cos(math.radians(ecliptic_longitude))

    a = math.atan(Y/X)

    right_ascension = 0
    if(X<0): 
        right_ascension = (a + math.pi)
    elif(X > 0 and Y < 0):
        right_ascension = (a + 2*math.pi)
    else:
        right_ascension = a

    declination = math.asin(math.sin(math.radians(obliquity))*math.sin(math.radians(ecliptic_longitude)))
    return np.array([math.degrees(right_ascension), math.degrees(declination)])

def eci_sun_vector(time):
    raw_sun_ra, sun_lat = raw_sun_vector(time)
    sun_lon = right_ascension_to_longitude(raw_sun_ra, time)

    ecef_sunx, ecef_suny, ecef_sunz = pymap3d.geodetic2ecef(sun_lat, sun_lon, 0)#vector that touches the surface
    sun_x, sun_y, sun_z = pymap3d.ecef2eci(ecef_sunx, ecef_suny, ecef_sunz, time)

    sun_vec = np.array([sun_x, sun_y, sun_z])
    sun_unit_vec = sun_vec/ (np.linalg.norm(sun_vec))
    return sun_unit_vec
    
```

```python
def expected_outputs(x, y, z, dx, dy, dz, qx, qy, qz, qtheta, dyaw, droll, dpitch):
    #x, y, z in eci meters, rot_quaternion should be initialized in RADIANS and should be a unit vector
    #rot_quaternion is the rotation quaternion for satellite w/r/t the ECEF frame
    rot_quaternion = Quaternion(axis = [qx, qy, qz], angle = qtheta)
    time = datetime.now()
    
    true_magnetic_vector = igdf_eci_vector(x, y, z, time)
    true_sun_vector = eci_sun_vector(time)
    observed_magnetic_vector = rot_quaternion.rotate(true_magnetic_vector)
    observed_sun_vector = rot_quaternion.rotate(true_sun_vector)
    
    return np.concatenate([observed_magnetic_vector, observed_sun_vector], axis = 0)
```

```python
class EKF():
    def __init__(self, f, h, q, r):
        self.f = f
        self.h = h
        self.R = r
        self.Q = q
        self.state_size = self.Q.shape[0]
        self.measurement_size = self.R.shape[0]

    def runthru(self, previous_state, received_outputs, covariance_mat):
        F = estimate_jacobian(self.f, previous_state)
        predicted_state = self.f(previous_state)
        predicted_cov = F@covariance_mat@(F.T) + self.Q

        H = estimate_jacobian(self.h, predicted_state)

        measurement_residual = received_outputs - self.h(previous_state)
        measurement_prediction_covariance = H@predicted_cov@(H.T) + self.R
        kalman_gain = predicted_cov@(H.T)@(np.linalg.inv(measurement_prediction_covariance))
    
        updated_state = predicted_state + kalman_gain@(measurement_residual)
        P = (np.eye(self.state_size) - kalman_gain@H)@predicted_cov
        return updated_state, P
```

```python
def dB_f(state):
    #state in the form [B, Bdot, Bdot2]
    dt = 1
    B = np.array([state[0], state[1], state[2]])
    Bdot = np.array([state[3], state[4], state[5]])
    Bdot2 = np.array([state[6], state[7], state[8]])
    l = np.concatenate([B + Bdot*dt, Bdot + Bdot2*dt, np.zeros(3)], axis = 0)
    return l

def dB_h(inputs):
    #state in the form [B, Bdot, Bdot2], searcy2012 paper & https://ieeexplore.ieee.org/document/9775141(somehow bypassed the paywall W)
    output = np.zeros(3)
    output[0] = inputs[0]
    output[1] = inputs[1]
    output[2] = inputs[2]
    return output

dbkf = EKF(dB_f, dB_h, q = np.eye(9), r = np.eye(3))
start = np.array([1000000, 0, 0, 0, 0, 0, 0, 0, 0])

cov = np.eye(9)
real = dB_h(np.array([1000000, 0, 0, 1000, 0, 0, 0, 0, 0]))
for i in range(100):
    last_real = real.copy()
    start, cov = dbkf.runthru(start,
                 real,
                 cov * 10)
    real[0] += (1000 + np.random.normal(scale = 10))
print(start)
```

```python
def omega(wx, wy, wz):
    return np.array([
        [0, -wx, -wy, -wz],
        [wx, 0, wz, -wy],
        [wy, -wz, 0, wx],
        [wz, wy, -wx, 0]
    ])

def simulate(x : np.float64, #eci x
             y: np.float64, #eci y
             z: np.float64, #eci z
             dx: np.float64, #velocity x
             dy: np.float64, #velocity y
             dz: np.float64, #velocity z
             dt: np.float64 #linear simulation rollout in seconds seconds
            ):
    #everything is in eci meters

    
    assumed_radial_distance = (6378+400)*1000 #m
    pos_vec = np.array([x,y,z], dtype = np.float64)
    velocity_vector = np.array([dx, dy, dz], dtype = np.float64 )
    acceleration_unit_vector = (-pos_vec) / (np.linalg.norm(pos_vec))

    G = 6.67428 * (10**(-11))
    mass = 5.9722 * (10**(24))
    acceleration_vector = (G*(mass)/(assumed_radial_distance**2)) * acceleration_unit_vector

    velocity_vector += acceleration_vector * dt
    pos_vec += (acceleration_vector * 1/2 * dt + velocity_vector * dt)
    
    return np.concatenate([pos_vec, velocity_vector],axis=0)
    
    
    
```

```python
def f(inputs):
    #scales everything to be in km
    x = inputs[0] * 1000
    y = inputs[1] * 1000
    z = inputs[2] * 1000
    dx = inputs[3] * 1000
    dy = inputs[4] * 1000
    dz = inputs[5] * 1000
    return simulate(x,y,z,dx,dy,dz, 0.1)/1000

def h(inputs):
    x = inputs[0] * 1000
    y = inputs[1] * 1000
    z = inputs[2] * 1000
    
    #The useful measurements used in this filter are the
    #measured magnitude of the Earth’s magnetic
    #field, and the measured dot product of the magnetic
    #field vector and the sun direction vector
    #https://www.sciencedirect.com/science/article/pii/S100093611160006X

    #seems to work decently so...
    outputs = expected_outputs(x,y,z, 0, 0, 0)
    magnetometer = np.array([outputs[0], outputs[1], outputs[2]])
    sun = np.array([outputs[3], outputs[4], outputs[5]])
    return np.concatenate([magnetometer, np.array([magnetometer@sun.T])], axis = 0)
```
